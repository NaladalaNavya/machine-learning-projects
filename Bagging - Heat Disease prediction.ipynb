when overfitting occurs we can do bagging and boosting

bagging:divides the data into (subsets of data)smaller sample data sets(m1,m2,m3) by picking random datapoints 
they may also select the some same data points randomly for different samples.and the model is getting trained on the that sample subsets of data and the predictions are madenow we consider the majority of the predictions or average for eg the prediction id 1,0,1 we consider as 1
this is also called as bootsrap aggregation.

```python
df5 = pd.get_dummies(df4,drop_first=True) #gets all the dummies for the data and drop the first category for each categorical variable to avoid multicollinearity 
df5.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
scores = cross_val_score(SVC(),X,y,cv= 5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
bagging = BaggingClassifier(base_estimator= SVC(),n_estimators = 100, max_samples=0.8,oob_score = True, random_state=0)
bagging.fit(X_train,y_train)
bagging.oob_score_
```

```python
bagging.score(X_test,y_test)
```
