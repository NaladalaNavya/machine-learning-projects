```python
import ast

# String representation of the list of dictionaries

# Convert the string to an actual list of dictionaries
result = ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
)

result
```

```python
def covert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i["name"])
  return l
```

```python
def covert(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!= 3:
      l.append(i["name"])
      counter += 1
  return l
```

```python
df["cast"] = df["cast"].apply(covert)
df["cast"]
```

```python
df.head()
```

```python
def fetch_director(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      l.append(i["name"])
      break
  return l
```

```python
df['crew']  = df['crew'].apply(fetch_director)
df['crew']
```

#bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 500,stop_words = "english")
vectors = cv.fit_transform(new_df["tags"]).toarray()
vectors
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
feature_names = cv.get_feature_names_out()
```

```python
feature_names
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
def stem(text):
  y =[]
  for  i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
ps.stem('danced')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df["title"]== movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
